# Pandas Export

- to sql

```python
import pandas as pd
df = pd.DataFrame()
```

# to_csv

```python
df = pd.read_csv('../datasets/dataset_session_20/deliveries.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv('batsman_runs.csv',index = False)
```

```python
df.pivot_table(index = 'batsman',columns = 'bowling_team',values = 'batsman_runs',aggfunc = 'sum')
```

# to_excel

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_excel('batsman_run.xlsx',sheet_name = 'Batsman_runs')
```

```python
temp_df_2 = df.pivot_table(index = 'batsman',columns = 'bowling_team',values = 'batsman_runs',aggfunc = 'sum')
```

```python
with pd.ExcelWriter('Output.xlsx') as writer:
    temp_df.to_excel(writer,sheet_name = 'Sheet_name1')
    temp_df_2.to_excel(writer,sheet_name = 'Sheet_name2')
```

# to_html

```python
df.columns
```

```python
df[df['batsman_runs']==6]
```

```python
df.query('batsman_runs == 6').pivot_table(index = 'over',columns = 'ball',values = 'batsman_runs',aggfunc = 'sum').to_html('sixes_heatmap.html')
```

# to_json

```python
df.head(2)
```

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('ipl.json')
```
