###`Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	  17	   10	       0	     20

```

````python
ipl.head(1)
````

````python
# code here
def count_win(group,team_number):
    group['win_team']= group[group[team_number] == group['WinningTeam']]['WinningTeam'].count()
    return group
def creating_groupby(group,teamnumber):
    new_dataframe_team1 = group.groupby(teamnumber)
    new_dataframe_team1 = new_dataframe_team1.apply(count_win,teamnumber)
    new_team1 = new_dataframe_team1[[teamnumber,'win_team']].drop_duplicates(keep = 'first')
    new_team1.set_index(teamnumber,inplace = True)
    return new_team1
def points(group):
    for index,row in group.iterrows():
        if row['no results'] == 0:
            group['points']= group['win']*2
        else:
            group['points'] = group['win']*2 + group['no results']*1
    return group
def point_table(group,season):
    new_dataframe = group[group['Season'] == season]
    new_dataframe_1 = pd.DataFrame(columns = ['Matches Played'])
    new_dataframe_1['Matches Played'] = new_dataframe.groupby('Team1')['MatchNumber'].count() + new_dataframe.groupby('Team2')['MatchNumber'].count()
    #################################################
    new_team1 = creating_groupby(new_dataframe,'Team1')
    new_team2 = creating_groupby(new_dataframe,'Team2')
    new_dataframe_1['win'] = new_team1 + new_team2
    ####################################################
    new_dataframe_1['no results']= new_dataframe[new_dataframe['WonBy'] == 'NoResults']['WonBy']
    new_dataframe_1['no results']= new_dataframe_1['no results'].fillna(0)
    ###################################################
    new_dataframe_1 = points(new_dataframe_1)
    return new_dataframe_1
final_data = point_table(ipl,'2022')
final_data.rename(columns = {'no results':'NoResult'},inplace = True)
final_data
````

````python
final_data.sort_index(inplace = True) 
````

````python
final_data.reset_index(inplace = True)
````

````python
final_data.head(2)
````

````python
final_data = final_data.sort_values('points',ascending = False)
````
