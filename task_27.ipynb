```python
final = pd.DataFrame()
for j in range(1,6):
    url = 'https://www.flipkart.com/search?q=smartphones&page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage = requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
#     company = soup.find_all('div',class_ = 'companyCardWrapper')
    phone = soup.find_all('div',class_ ='_2kHMtA')

    phone_image_url = []
    phone_name = []
    avg_ratings = []
    total_ratings = []
    total_reviews = []
    discounted_price = []
    actual_price = []
    for i in phone:
        try:
            phone_image_url.append(i.find('img',class_ ='_396cs4')['src'])
        except:
            phone_image_url.append(np.nan)
        try:
            phone_name.append(i.find('div',class_='_4rR01T').text.strip())
        except:
            phone_name.append(np.nan)
        try:
            avg_ratings.append(i.find('div',class_='_3LWZlK').text.strip())
        except:
            avg_ratings.append(np.nan)
        try:
            total_ratings.append(i.find_all('span',class_ = '_2_R_DZ')[0].find_all('span')[0].find_all('span')[0].text.strip())
        except:
            total_ratings.append(np.nan)
        try:
            total_reviews.append(i.find_all('span',class_ = '_2_R_DZ')[0].find_all('span')[0].find_all('span')[2].text.strip())
        except:
            total_reviews.append(np.nan)
        try:
            discounted_price.append(i.find('div',class_='_30jeq3').text.strip())
        except:
            discounted_price.append(np.nan)
        try:
            actual_price.append(i.find('div',class_='_3I9_wc _27UcVY').text.strip())
        except:
            actual_price.append(np.nan)
    d = pd.DataFrame({'Phone_Image_Url':phone_image_url,'Phone_name':phone_name,'average_ratings':avg_ratings,'Total_ratings':total_ratings,'Total_reviews':total_reviews,'discounted_Price':discounted_price,'Actual_Price':actual_price})
    final = final.append(d,ignore_index = True)
```
